## Fonctions de base

```python
%pylab inline
import numpy as np
import pylab

def calcul_perte_moyenne(w, b, data, classes, c, n):
    sum = 0
    for i in range(n):
        if (np.dot(w, data[i]) + b) * classes[i] < 1:
            sum += 1 - (np.dot(w, data[i]) + b) * classes[i] + (np.linalg.norm(w) ** 2) / (2 * n * c)
        else:
            sum += (np.linalg.norm(w) ** 2) / (2 * n * c)
    perte = sum / n
    return perte

def calcul_taux_erreur(w, b, data, classes, n):
    count = 0
    for i in range(n):
        if np.sign(np.dot(w, data[i]) + b) == classes[i]:
            count += 1
    taux = (1 - float(count) / n) * 100.0
    return taux

def descente(svm):
    index = 0
    count = 0
    while count < svm.m:
        pass
        svm.theta += -svm.eta * gradient(svm, svm.theta, index)
        index += svm.k
        if index >= svm.num_train:
            index = index % svm.num_train
            count += 1
            perte = calcul_perte_moyenne(svm.theta[:svm.dim_train], svm.theta[-1], svm.train_x, svm.classes, svm.c, svm.num_train)
            taux = calcul_taux_erreur(svm.theta[:svm.dim_train], svm.theta[-1], svm.train_x, svm.classes, svm.num_train)
            print ''
            print 'Époque', count
            print 'perte moyenne :', perte
            print 'taux d\'erreur :', taux,'%'
#           decommenter pour obtenir les regions de decision a chaque epoque
#             gridplot(svm, svm.train_data, svm.train_data, 50, select_method=np.sign)
    return svm.theta
    
def gradient(svm, theta, index):
    grad = np.zeros(svm.dim_train+1)
    for i in range(index, index + svm.k + 1):
        l = i % svm.num_train
        if (np.dot(theta[:svm.dim_train], svm.train_x[l,:]) + theta[-1]) * svm.classes[l] < 1:
            for j in range(svm.dim_train):
                grad[j] += theta[j] / (svm.k * svm.c) - svm.train_x[l,j] * svm.classes[l]
            grad[-1] += -svm.classes[l]
        else:
            for j in range(svm.dim_train):
                grad[j] += theta[j] / (svm.k * svm.c)
    grad /= svm.k
    return grad
```

## Algorithme de SVM linéaire par descente de gradient minibatch

```python
class svm:
    def __init__(self, c, k, m, eta, poly_deg = None):
        self.c = c
        self.k = k
        self.m = m
        self.eta = eta
        self.poly_deg = poly_deg

    def train(self, train_data):
        self.train_data = train_data
        self.num_train = train_data.shape[0]
        self.classes = train_data[:,-1]
        self.train_ex = train_data[:,:-1]
        self.dim_train = self.train_x.shape[1]
        self.theta = np.zeros(self.dim_train+1)
        self.theta = descente(self)

    def compute_predictions(self, test_data):
        num_test = test_data.shape[0]
        resultats = np.zeros((num_test,1))
        for (i,ex) in enumerate(test_data):
            if(self.dim_train != ex.shape[0]):
                ex = transformation_non_lineaire(ex,self.poly_deg)
            y = np.dot(self.theta[:self.dim_train], ex) + self.theta[-1]
            if y > 0:
                resultats[i,0] = 1
            if y < 0:
                resultats[i,0] = -1
        return resultats
```

## Fonctions d'affichage graphique

```python
# fonction plot
def gridplot(classifieur,train,test,n_points=50, select_method = lambda X: numpy.argmax(X,axis=1)):
    train_test = numpy.vstack((train,test))
    (min_x1,max_x1) = (min(train_test[:,0]),max(train_test[:,0]))
    (min_x2,max_x2) = (min(train_test[:,1]),max(train_test[:,1]))

    xgrid = numpy.linspace(min_x1,max_x1,num=n_points)
    ygrid = numpy.linspace(min_x2,max_x2,num=n_points)

    # calcule le produit cartesien entre deux listes
    # et met les resultats dans un array
    thegrid = numpy.array(combine(xgrid,ygrid))

    resultats = classifieur.compute_predictions(thegrid)

    classesPred = select_method(resultats)

    # La grille
    # Pour que la grille soit plus jolie
    #props = dict( alpha=0.3, edgecolors='none' )
    pylab.scatter(thegrid[:,0],thegrid[:,1],c = classesPred, s=50)
    # Les points d'entrainment
    pylab.scatter(train[:,0], train[:,1], c = train[:,-1], marker = 'v', s=150)
#     Les points de test
#     pylab.scatter(test[:,0], test[:,1], c = test[:,-1], marker = 's', s=150)

    ## Un petit hack, parce que la fonctionalite manque a pylab...
    h1, = pylab.plot([min_x1], [min_x2], marker='o', c = 'w',ms=5) 
    h2, = pylab.plot([min_x1], [min_x2], marker='v', c = 'w',ms=5) 
    h3, = pylab.plot([min_x1], [min_x2], marker='s', c = 'w',ms=5) 
    handles = [h1,h2,h3]
    ## fin du hack

    labels = ['grille','train','test']
    pylab.legend(handles,labels)

    pylab.axis('equal')
    pylab.show()

## http://code.activestate.com/recipes/302478/
def combine(*seqin):
    '''returns a list of all combinations of argument sequences.
for example: combine((1,2),(3,4)) returns
[[1, 3], [1, 4], [2, 3], [2, 4]]'''
    def rloop(seqin,listout,comb):
        '''recursive looping function'''
        if seqin:                       # any more sequences to process?
            for item in seqin[0]:
                newcomb=comb+[item]     # add next item to current comb
                # call rloop w/ rem seqs, newcomb
                rloop(seqin[1:],listout,newcomb)
        else:                           # processing last sequence
            listout.append(comb)        # comb finished, add to list
    listout=[]                      # listout initialization
    rloop(seqin,listout,[])         # start recursive process
    return listout
```

## Transformation linéaire

```python
def pre_traitement_non_lineaire(data, k):
    newData = None
    for i in range(data.shape[0]):
        if(newData is None):
            newData =  transformation_non_lineaire(data[i], k)
        else:
            newData = np.vstack((newData, transformation_non_lineaire(data[i], k)))
    return newData
```

```python
def transformation_non_lineaire(x, k):
    assert k>1 and k<5
    if(k == 2):
        ncaract = 5
    elif(k == 3):
        ncaract = 9
    elif(k == 4):
        ncaract = 14
    newLine = np.empty(ncaract)
    newLine[0] = x[0]
    newLine[1] = x[1]
    newLine[2] = x[0]*x[1]
    newLine[3] = x[0]**2
    newLine[4] = x[1]**2
    if(k >= 3):
        newLine[5] = x[0]**2 * x[1]
        newLine[6] = x[0] * x[1]**2
        newLine[7] = x[0]**3
        newLine[8] = x[1]**3
    if(k >= 4):
        newLine[9] = x[0]**3 * x[1]
        newLine[10] = x[0]**2 * x[1]**2
        newLine[11] = x[0] * x[1]**3
        newLine[12] = x[0]**4
        newLine[13] = x[1]**4
    return newLine
```

```python
train_targets = train_set[:,-1]
test_targets = train_set[:,-1]
poly_deg = 4
#Ensembles transformés sans étiquette.
transformed_train_inputs = pre_traitement_non_lineaire(train_set[:,:-1], poly_deg)
transformed_test_inputs = pre_traitement_non_lineaire(test_set[:,:-1], poly_deg)
```

```python
#Ensembles transformés avec étiquette.
transformed_train = np.column_stack((transformed_train_inputs, train_targets))
transformed_test = np.column_stack((transformed_test_inputs, test_targets))
```

```python
# Taille de C
print "On va entraîner une SVM avec un C de",c,"et",m,"époques sur",n_train,"exemples d'entraînement avec transformation polynomiale explicite avec un degré de",poly_deg,"."

# Creer le classifieur
model = svm(c, k, m, eta, poly_deg)
# L'entrainer
model.train(transformed_train)

# Surface de decision
gridplot(model,transformed_train,transformed_test,n_points = grid_size, select_method=np.sign)
```
